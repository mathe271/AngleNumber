```python
import math
import pandas as pd
import numpy as np
import re
```

```python
def primeFactor(N):
    i = 2
    flg = True
    PF = []
    
    while i <= math.sqrt(N)+1:
        
        if N % i == 0:
            PF.append(i)
            N = N // i
            flg = False
        else:
            i += 1
            
    if flg == True:
        PF.append(N)
    
    return PF
```

```python
PFList = []

for i in range(1,367):
    pf = primeFactor(i)
    PFList.append([i,pf,len(pf),list(set(pf)),len(list(set(pf)))])
```

```python
df = pd.DataFrame(PFList,columns=['number','PF','PF_Length','dedupPF','dedupPF_Length'])
```

```python
df.set_index("number",inplace=True)
```
